```python
user_symptoms = [
"I have red patches of skin that look inflamed and are covered with thick white scales. They appear mostly on my lower back and elbows. The affected areas are dry and itchy, and sometimes the skin cracks open."
"For several days now, I’ve been experiencing a high fever, chills, and a sharp pain in my chest when I breathe in. I feel weak and extremely tired, and my cough keeps getting worse."
"I have oily skin and recently developed small red bumps and blackheads on my nose and chin. There are also a few painful cysts around my jawline."
]

symptoms_review = "\n".join([f"Symptom {i+1}: {review}" for i,
review in enumerate(user_symptoms)])

parameters = {
"top_k": 5,
"top_p": 0.9,
"max_tokens": 64,
"min_tokens": 0,
"random_seed": None,
"repetition_penalty": 1.0,
"stopping_criteria": "length (64 tokens)",
"stopping_sequence": None
}
```

```python
# Add initial prompt
refined_prompt = f""":
Classify each of the following symptoms into the most likely disease.
Return only the disease name for each symptom.
{symptoms_review}
"""
# Invoke the model
response = output.invoke(refined_prompt, parameters=parameters)
# Print the response
print("Granite Model Refined Response:\n")
print(response)
```
